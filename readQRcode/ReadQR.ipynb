# <font style="color:rgb(50,120,229)">QR reader</font>

Read an image that contains a QR code. 

`image`: Image on which the QR is to be found.



```python
# Import cv2 module
import cv2

# Import matplotlib for displaying image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Path to the image we are going to read
imageName = "info.jpg"
#imageName = "taco.jpg"

# Load the image
image = cv2.imread(imageName, cv2.IMREAD_COLOR)
```

```python
# Display the original image
plt.imshow(image[:,:,::-1])
```

### <font style="color:rgb(8,133,37)">Find the QR</font>
Find the location of the QR code and read the contents.

```python
# Get an instance of QRCodeDetector, and run with our image
detector = cv2.QRCodeDetector()
decodedText, cornerPoints, _ = detector.detectAndDecode(image)
```

```python
imageWithQR = image.copy()
#draw a rectangle around the QR code, if found

if cornerPoints is not None:
    # QR Code detected handling code
        print("QR code detected")
        print(decodedText)    
        #Corners of the bounding box
        pt1 = (int(cornerPoints[0][0][0]), int(cornerPoints[0][0][1]))
        pt2 = (int(cornerPoints[0][2][0]), int(cornerPoints[0][2][1]))
        cv2.rectangle(imageWithQR, pt1, pt2, (0, 255, 0), thickness=4, lineType=cv2.LINE_AA)
else:
        print("QR code not detected")
        cv2.putText(imageWithQR, 'QR not found', (0,100),cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2, cv2.LINE_AA)

```

```python
# Display the image
plt.imshow(imageWithQR[:,:,::-1])
```
